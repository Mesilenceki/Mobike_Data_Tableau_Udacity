```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('mobike_shanghai_sample_updated.csv')
```

```python
df_1.head()
```

```python
df = df_1.track.str.split(pat = '#',expand=True)
```

```python
df = pd.concat([df_1[["orderid","bikeid","userid","start_time","end_time","start_location_x","start_location_y","end_location_x","end_location_y"]],df],axis = 1)
df = df.melt(id_vars=["orderid","bikeid","userid","start_time","end_time","start_location_x","start_location_y","end_location_x","end_location_y"],var_name = 'track_id')
```

```python
df = df[df["value"].notna()]
```

```python
df
```

```python
data = df['value'].str.split(",", expand=True)
data.columns = ['longitude','latitude']
df = pd.concat([df, data], axis=1)
```

```python
df
```

```python
df.drop(['value'], axis=1, inplace=True)
```

```python
df["start_time"] = pd.to_datetime(df["start_time"])
df["end_time"] = pd.to_datetime(df["end_time"])
```

```python
df['duration']= pd.to_datetime(df.end_time)-pd.to_datetime(df.start_time)
df['duration']= df.duration.apply(lambda x:round(x.total_seconds()/60))
```

```python
df
```

```python
df.to_csv("track_data.csv")
```
